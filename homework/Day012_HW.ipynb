### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "lxml")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find('a').text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
tags = {}

for d in soup.find(class_="part_list_2").find_all('h3'):
    date = d.find(class_="date").text
    title = str(d.find('a').text).replace('\u3000','')
    tag = d.find('em').text
#     tag = d.select('.tag')[0].text
    tags.setdefault(tag, [])
    tags[tag].append({
        'title': title,
        'date' : date
    })

print(tags)
```

### ③ 哪一個類別的文章最多

```python
data = []
for i in tags:
    print(i,len(tags[i]))
    d = {}
    d['tag'] = i
    d['count'] = len(tags[i])
    data.append(d)

for d in sorted(data, key=lambda d: d['count'],reverse=True):
    print(d)
```
