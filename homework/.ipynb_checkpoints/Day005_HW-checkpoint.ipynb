```python
import requests,json

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = json.loads(r.text)
```

```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

print('é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³ {} ç­†è³‡æ–™'.format(len(response)))
print('æ¯ä¸€ç­†è³‡æ–™åŒ…å«ä¸‹åˆ—æ¬„ä½:')
print(response[0].keys())
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

for data in response:
    print('æ¨™é¡Œ: ' + data['title'])
    print('è²¼æ–‡æ™‚é–“: ' + data['createdAt'])
    print('ç•™è¨€äººæ•¸: ' , data['commentCount'])
    print('æŒ‰è®šäººæ•¸: ' , data['likeCount'] , '\n')
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

comment_sum = 0
like_sum = 0

for count in response:
    comment_sum += count['commentCount']

for count in response:
    like_sum += count['likeCount']


print('ç†±é–€æ–‡ç« çš„å¹³å‡ç•™è¨€äººæ•¸: ', comment_sum/len(response))
print('ç†±é–€æ–‡ç« çš„å¹³å‡æŒ‰è®šäººæ•¸: ', like_sum/len(response))
print()

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
response = json.loads(r.text)

comment_sum = 0
like_sum = 0

for count in response:
    comment_sum += count['commentCount']

for count in response:
    like_sum += count['likeCount']

print('éç†±é–€æ–‡ç« çš„å¹³å‡ç•™è¨€äººæ•¸: ', comment_sum/len(response))
print('éç†±é–€æ–‡ç« çš„å¹³å‡æŒ‰è®šäººæ•¸: ', like_sum/len(response))
```
