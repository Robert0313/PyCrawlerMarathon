# Ettoday 網路爬蟲實作練習


* 能夠利用 Request + BeatifulSour 撰寫爬蟲，並存放到合適的資料結構


## 作業目標

根據範例：

1. 取出今天所有的新聞
2. 取出現在時間兩小時內的新聞
3. 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import datetime
from dateutil.parser import parse

import time

url = 'https://www.ettoday.net/news/news-list.htm'
```

## 1. 取出今天所有的新聞

```python
datetime_today = datetime.datetime.today()
datetime_today = datetime_today.strftime("%Y/%m/%d")
datetime_today
```

```python
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get(url)
```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'lxml')

d = soup.find(class_="part_list_2").find('h3')
checkDate = parse(d.find(class_="date").text).strftime("%Y/%m/%d")

while checkDate == datetime_today:
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, 'lxml')
    d = soup.find(class_="part_list_2").find_all('h3')[-1]
    checkDate = parse(d.find(class_="date").text).strftime("%Y/%m/%d")
    time.sleep(2)
```

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
    checkDate = parse(d.find(class_="date").text).strftime("%Y/%m/%d")
    if checkDate == datetime_today:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

```python
browser.close()
```

## 2. 取出現在時間兩小時內的新聞

```python
datetime_today = datetime.datetime.today()
time_delta = datetime.timedelta(hours=-2)
new_dt = datetime_today + time_delta
new_dt = new_dt.strftime("%Y/%m/%d %H:%M:%S")
new_dt = parse(new_dt)

print(new_dt)
print(datetime_today)
```

```python
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get(url)
```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'lxml')

d = soup.find(class_="part_list_2").find('h3')
checkDate = parse(d.find(class_="date").text)

for d in soup.find(class_="part_list_2").find_all('h3'):
    checkDate = parse(d.find(class_="date").text)
    if new_dt <= checkDate <= datetime_today:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
    else:
        break
```

```python
browser.close()
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
datetime_today = datetime.datetime.today()
time_delta = datetime.timedelta(days=-3)
new_dt = datetime_today + time_delta
new_dt = new_dt.strftime("%Y-%#m-%#d")

startTime = parse(new_dt).replace(hour=15)
endTime = startTime.replace(hour=17)

print(startTime)
print(endTime)
```

```python
new_url = 'https://www.ettoday.net/news/news-list' + '-' + new_dt + '-0' + '.htm'
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get(new_url)
```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'lxml')

d = soup.find(class_="part_list_2").find('h3')
checkDate = parse(d.find(class_="date").text)

while checkDate >= startTime:
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, 'lxml')
    d = soup.find(class_="part_list_2").find_all('h3')[-1]
    checkDate = parse(d.find(class_="date").text)
    time.sleep(2)
```

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
    checkDate = parse(d.find(class_="date").text)
    if startTime <= checkDate <= endTime:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

```python
browser.close()
```
